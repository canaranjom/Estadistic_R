```r
datos <- data.frame(
  Edad = c(23, 23, 24, 25, 26, 27, 28, 30, 32, 34),
  experiencia_laboral = c(1, 1.5, 2, 3, 2, 6, 8, 8.5, 10, 12)
)

print(datos)
```

```r
plot(datos, xlab = "Edad", ylab = "experiencia_laboral")
```

```r
modelo<-lm(experiencia_laboral~Edad,data=datos)
summary(modelo)
```

Valor B0 = 22.25466, B1 = 0.91580
